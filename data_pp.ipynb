```python
import pandas as pd

import os

def process_plt_file(file_path):
    # Read the .plt file into a DataFrame, skipping the first 6 lines of metadata
    try:
        df = pd.read_csv(file_path, sep='\s+', header=None, skiprows=6)
        return df
    except Exception as e:
        print(f"Failed to process {file_path}: {e}")
        return None

def process_directory(root_dir):
    all_data = []

    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.plt'):
                folder_name = os.path.basename(os.path.dirname(root))

                try:
                    # Extract folder number (000 to 181)
                    folder_number = int(folder_name)
                except ValueError:
                    print(f"Skipping folder {folder_name}: not a valid number")
                    continue

                file_path = os.path.join(root, file)
                print(f"Processing {file_path}")

                df = process_plt_file(file_path)
                if df is not None:
                    df['Folder_Number'] = folder_number  # Add a column for the folder number
                    all_data.append(df)

    # Concatenate all DataFrames into one
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        return final_df
    else:
        return None

if __name__ == "__main__":
    directory_path = 'C:\Projects\Load_Balancing\Geolife Trajectories 1.3'  # Replace with the path to your directory
    final_df = process_directory(directory_path)

    if final_df is not None:
        # print(final_df.head())  # Display the first few rows of the final DataFrame

        final_df.to_csv("Test.csv")
        # Save to a CSV file if needed
        # final_df.to_csv('combined_data.csv', index=False)
    else:
        print("No data processed.")
```

```python
import csv

# Function to process each row
def process_row(row):
    # Discard the first column
    _, second_col, third_col = row

    # Split the second column into separate fields
    fields = second_col.split(',')

    # Combine the extracted fields with the third column
    result = fields + [third_col]

    return result

# Function to read, process, and save the CSV file
def read_process_save_csv(input_file_path, output_file_path):
    processed_data = []

    # Open the input CSV file for reading
    with open(input_file_path, mode='r') as infile:
        csv_reader = csv.reader(infile)

        # Process each row in the input file
        for row in csv_reader:
            processed_row = process_row(row)
            processed_data.append(processed_row)

    # Open the output CSV file for writing
    with open(output_file_path, mode='w', newline='') as outfile:
        csv_writer = csv.writer(outfile)

        # Write each processed row to the output file
        csv_writer.writerows(processed_data)

# Example usage
input_file_path = 'Test.csv'  # Replace with your actual input file path
output_file_path = 'output_file.csv'  # Replace with your desired output file path
read_process_save_csv(input_file_path, output_file_path)
```
